*This tutorial will serve as a practical exercise for familiarizing oneself with the Matplotlib library.*

## 1- Import the Libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
```

## 2- Read the DataSet

```python
data_frame=pd.read_csv(r"Jupyter DataSets/TSLA_DataSet.csv")
```

## 3- Represent the Date and open price in Plot 

```python
plt.plot(data_frame.Date,data_frame.Open)
```

## 4- Add Labels and title to plot 

```python
plt.ylabel("Open Price")
plt.xlabel("Date")
plt.title("TSLA")
plt.plot(data_frame.Date,data_frame.Open)
```

## 5- Change The Color 

```python
plt.ylabel("Open Price")
plt.xlabel("Date")
plt.title("TSLA")
plt.plot(data_frame.Date,data_frame.Open,color="green")
```

## 6- Add two values (open and close ) in the Plot 

```python
plt.ylabel("Open and Close Price")
plt.xlabel("Date")
plt.title("TSLA")
plt.plot(data_frame.Date,data_frame.Open,label="Open")
plt.plot(data_frame.Date,data_frame.Close,label="Close")
plt.grid()
plt.legend()
```

## 7-Same example with Bar Chart 

```python
plt.ylabel("Open Price")
plt.xlabel("Date")
plt.title("TSLA")
plt.bar(data_frame.Date,data_frame.Open,width=0.5)
```

## 8- Histogram 

```python
plt.hist(data_frame.Open,rwidth=0.92)
```

## 9- Histogram For 2 Columns

```python
plt.hist([data_frame.Open,data_frame.Close],rwidth=0.92,label=["Open","Close"])
plt.legend()
```

## 10- Save plot as an Image

```python
plt.hist([data_frame.Open,data_frame.Close],rwidth=0.92,label=["Open","Close"])
plt.legend()
plt.savefig("tsla hist open and close.png")
```
